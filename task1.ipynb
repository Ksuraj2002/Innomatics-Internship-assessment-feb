```python
#task 2 pass/fail analyser
```

```python
passcnt = 0
failcnt = 0
marks = [45,78,90,33,60]

for mark in marks:
    if mark >= 50:
        passcnt+=1
    else:
        failcnt+=1
print("results:")
print(f"Total no of students passed = {passcnt}" )
print(f"Total no of students failed = {failcnt}")
```

```python
names = [" Alice ", "bob", " CHARLIE "]
ans = []
for name in names:
    temp = name.strip()
    temp2 = temp.lower()
    ans.append(temp2)

print(f"{ans}")
```

```python
#task 4 message length analyser
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    message_length = len(message)
    print(f"Message: '{message}' | Length: {message_length}", end="")
    if message_length > 10:
        print("long message")
    else:
        print("")
```

```python
#task 5 error message detector
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
error_keyword = "ERROR"

for log_entry in logs:
    if log_entry == error_keyword:
        error_count += 1

print("Error Message Detector Results:")
print(f"System logs: {logs}")
print(f"Total '{error_keyword}' entries detected: {error_count}")
```
