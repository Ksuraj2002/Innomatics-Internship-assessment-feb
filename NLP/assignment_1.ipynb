# Task 2

```python
import re

def preprocess_text(text):
    if not text: return ""
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs and Emails 
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 4. Handle repeated characters 
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 5. Remove special characters 
    text = re.sub(r'[^\w\s]', '', text)
    
    # 6. Filter length and remove extra spaces 
    tokens = text.split()
    exceptions = {'no', 'not'}
    cleaned_tokens = [t for t in tokens if len(t) > 2 or t in exceptions]
    
    return " ".join(cleaned_tokens).strip()
```

# Task 4

```python
def get_token(cleaned_sentence):
    
    tokens = cleaned_sentence.split()
    
    if not tokens:
        return 0, 0, 0.0
    
    total_tokens = len(tokens)
    
    unique_tokens = len(set(tokens))
    
    avg_length = sum(len(word) for word in tokens) / total_tokens
    
    return total_tokens, unique_tokens, round(avg_length, 2)
```

# Task 5 and 6

```python
from collections import Counter

def full_pipeline(text_list):
    clean_sentences = [preprocess_text(s) for s in text_list]
    all_tokens = [t for s in clean_sentences for t in s.split()]
    
    # Task 5: Frequency Analysis 
    counts = Counter(all_tokens)
    top_10 = counts.most_common(10)
    least_5 = counts.most_common()[:-6:-1]
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences,
        "top_10": top_10,
        "least_5": least_5
    }
```

```python
text_list = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
full_pipeline(text_list)
```
