# Soln 1(Smart Parking Lot Management System)

```python
n = int(input("Enter Parking Capacity"))
arr = list(input().split())

cnt = 0

for i in arr:
    if cnt==n:
        break
    if i == "IN":
        cnt+=1
    else: cnt-=1

print("Currently Parked Vehicles: ",cnt)

if cnt==n:
    print("Parking Status: Unavailable")
else :
    print("Parking Status: Available")

```

# Soln 2(Online Food Delivery Time Estimator)

```python
def calculate_time(dis,wc,tl):
    time = dis*5
    if wc == "Rainy":
        time += 10
    if tl == "high":
        time += 5
    return time

dis = int(input("Enter distance in km: "))
tl = input("Enter traffic level: ")
wc = input("Enter weather condition: ")

time = calculate_time(dis,wc,tl)
print(time)
    
```

# Soln 3 (Movie theatre seat occupancy analyzer)

```python
def calculate_seat(ts,arr):
    cnt = 0
    for i in arr:
        cnt += i

    per = (cnt/ts)*100
    return per

ts = int(input("Enter total number of seats: "))
arr = list(map(int,input("Enter the list containing seat occupancy: ").split()))
per = calculate_seat(ts,arr)
print("Occupancy: ",per,"%")

if per>=75:
    print("Almost Full")
else : print("Seats Available")
```

# Soln 4(Cloud Server Load Classification System)

```python
def server_load(arr):
    num = 0
    for i in arr:
        num += i

    avg = int(num/len(arr))
    print("Average CPU Load: ",avg,"%")

    if avg<50:
        print("Server Status: Normal")
    elif avg<=80:
        print("Server Status: Warning")
    else: print("Server Status: Critical")

arr = list(map(int,input("Enter CPU Readings in %: ").split()))
avg = server_load(arr)
```

# Soln 5 : Smart Classroom Resource Usage Monitor

```python
def usage_check(p,a,l):
    ans = []
    if p>8:
        ans.append("Projector")
    if a>8:
        ans.append("AC")
    if l>8:
        ans.append("Lights")
    return ans

p = int(input("Enter projector usage in hours: "))
a = int(input("Enter AC usage in hours: "))
l = int(input("Enter Lights usage in hours: "))
ans = []
ans = usage_check(p,a,l)
print("Overused Resources: ",ans)
if(len(ans) > 0) :
    print("Energy Alert: Yes ")
else : print("Energy Alert: No")
```

# Soln 6: Online Event Registration Capacity Controller

```python
def event(cap,reg):
    confirm = min(cap,reg)
    print("Confirmed Registrations: ",confirm)
    wait = max(0,reg-cap)
    print("Waitlisted Users: ",wait)
    if reg>=cap:
        print("Registration Status: Closed")
    else: print("Registration Status: Open")

cap = int(input("Enter Event Capacity: "))
reg = int(input("Enter number of registrations: "))
event(cap,reg)
```
